## Background and Prior Work


During an NBA game, each team sends five players to battle it out on the court, showcasing their skills and strategy to determine who’s better. But what about the invisible sixth man in the arena: the fans? Beyond the players and coaches, does the energy and sheer presence of fans affect a team’s performance? And in turn, does a team’s success on the court bring more fans into the stands? We will set out to cover this symbiotic relationship between team performance and attendance numbers by aiming to answer the question of if an NBA team’s performance will influence the number of attendants during the 2001-2023 seasons. For this study, performance refers to the team's wins in both home and away games.

We have found this to be an interesting topic as fans of the NBA since it would be insightful to see if there is correlation between how good a team is and how many fans there are to witness the games in person. Outside of our own love for the game, we figure that this research project will help answer other questions regarding the NBA such as if home court advantage matters, how a team’s players and marketability of those players affect attendance, and how a team’s attendance numbers can fluctuate over time due to how a team is performing. 

In a previous study conducted by Chenzao Wang in 2024 called “The Effect of NBA Attendance on Team’s Performance”<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), they covered the impact of attendance on team performance during the 2020-2021 NBA season in comparison to regular NBA seasons. The 2020-2021 was the season that got hit by COVID-19 complications in the middle of the playoffs, moving 22 NBA teams to closed arenas in Orlando, Florida, dubbed the “Bubble” era of the NBA. During this time audiences were very limited (only consisting of close friends, family, and staff), and there wasn’t really a home-court advantage as every team played in the same arena in Orlando (there was only one team that could have potentially benefited from “home court advantage”, the Orlando Magic). Wang took the stats of all 30 NBA teams from 2015-2022 (Field Goal Percentage, Free Throws Made, etc.) and the attendance numbers of the 2020-2021 “Bubble” season and analyzed them to see if there was a difference in team performance when there were fans in attendance and little to no fans in attendance. The study concluded in finding there was a slight advantage to home teams when they had audiences present (normal NBA seasons), and home court advantage was less obvious when there were no audiences (COVID season). This finding supports the theory that NBA team performance is positively affected by the fans in attendance. 

The influence of star players on NBA attendance has also been well-documented as Metcalfe (2013)<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) explored the impact of "star power" on fan attendance in his paper "NBA Star Power: Impact on Attendance." This study examined how having highly skilled or well-known players on a team affects fan turnout, analyzing data over several seasons. Metcalfe’s findings suggested that star players are a significant driver of attendance, as fans are more likely to attend games featuring popular or top-performing athletes. This concept of "star power" extends beyond wins and losses, suggesting that fans often view games as an opportunity to see high-profile athletes perform, regardless of the team’s overall record. By examining this factor, Metcalfe’s study sheds light on how audience patterns might align with elements outside of win-loss records, offering a framework for our research as we investigate how team success (wins across both home and away games) relates to attendance across a longer timeline. 

Previous studies used logistic regression models and the Bradley-Terry model to analyze performance and attendance relationships. Wang's work, for instance, utilized logistic models to assess how attendance influences home-court advantage. In our analysis, we plan to use a similar approach by leveraging statistical tools such as regression modeling to find out the relationship between performance metrics and attendance over a broader timeline. We will also account for star presence and team market size to create an accurate model for our research question. These methods will allow us to explore how team performance and other factors influence fan engagement, which provides valuable insights onto NBA attendance patterns from 2001 to 2023.

# Data Cleaning

While we have all the necessary data, it is spread across three separate datasets. To effectively merge them, we must first clean and standardize each dataset, ensuring consistent variable names to facilitate a seamless integration process.

## df_nba_standings Cleaning

In the df_nba_standings dataset, there are unnecessary columns like "Rk" that provides data that isn't directly relevant to our analysis. In order to make the dataset more compact and easier to read, we will drop those columns first. In addition to that, the combined columns (e.g. "Overall", "Home", "Road") contain both wins and losses in a single string format (e.g. 61-21). Splitting these into separate Wins and Losses columns ensures better usability for numerical analysis, enabling us to calculate metrics or create visualizations without additional parsing. After extracting the relevant data (wins and losses), the original columns are redundant and clutter the dataset so we drop those columns as well. We also standardize column names (e.g. converting to lowercase) to ensure consistency and avoid errors during merging or further processing, especially in case-sensitive environments.

```python
# Drop unnecessary columns
df_nba_standings = df_nba_standings.drop(['Rk', 'E', 'W', 'Pre', 'Post', '?3', '?10'], axis = 1)

# Split Overall, Home, and Road columns into Wins and Losses
df_nba_standings[['Overall Wins', 'Overall Losses']] = df_nba_standings['Overall'].str.split('-', expand=True).astype(int)
df_nba_standings[['Home Wins', 'Home Losses']] = df_nba_standings['Home'].str.split('-', expand=True).astype(int)
df_nba_standings[['Road Wins', 'Road Losses']] = df_nba_standings['Road'].str.split('-', expand=True).astype(int)

# Drop the original Overall, Home, and Road columns
df_nba_standings = df_nba_standings.drop(columns=['Overall', 'Home', 'Road'])

# Lowercase all column names
df_nba_standings.columns = df_nba_standings.columns.str.lower()
```

Historically, some teams have changed names or relocated over the years (e.g. New Jersey Nets to Brooklyn Nets). So, we decided to standardize team names ensure consistency, preventing duplication or confusion when merging datasets.

```python
# Create list of standardized team names
team_name_mapping_refined = {
    "New Jersey Nets": "Brooklyn Nets",
    "Seattle SuperSonics" : "Oklahoma City Thunder",
    "Charlotte Bobcats" : "Charlotte Hornets",
    "New Orleans Hornets" : "New Orleans Pelicans",
    "New Orleans/Oklahoma City Hornets" : "New Orleans Pelicans"
}

# Apply the refined mapping to standardize the team names
df_nba_standings['team'] = df_nba_standings['team'].replace(team_name_mapping_refined)
```

```python
# Refine team names
df_nba_standings.loc[10, 'team'] = "New Orleans Pelicans"
df_nba_standings = df_nba_standings.drop(index=114)
```

## Merge Datasets

The next step is to merge all three datasets into a single, unified dataset. By combining the datasets, we can analyze all relevant information together, allowing for a more comprehensive exploration of the data. This will enable us to examine relationships between team performance, attendance, and Elo ratings across different seasons, ensuring consistency and making the analysis more effective. ### Distribution of Wins

First we displayed the distribution of the overall wins of all NBA teams from 2001-2023 using histplot. 

```python
sns.histplot(data= clean_dataset['overall_wins'], bins = 30, kde = True)
plt.xlabel('Overall Wins from 2001-2023')
```

The histogram displays the distribution of overall wins across all teams during this time period. It was revealed that the distribution is seemingly normal.

### Home vs Away (Road) Wins

Now we want to examine the overall wins across the different types of games: home and away (road).

Home games can often provide teams with an advantage due to supportive crowds and familiar settings, whereas road games can be more challenging due to strenuous travel and unfamiliar environments. We used a KDE plot to compare the distribution of wins for home and road games, highlighting differences in performance based on game location 

```python
sns.kdeplot(data= clean_dataset, x = 'home_wins', label = 'Home')
sns.kdeplot(data= clean_dataset, x = 'road_wins', label = 'Road')

plt.title('KDE Plot for Home, and Road Wins')
plt.xlabel('Wins')
plt.legend(title='Win Type')
plt.tight_layout()
plt.show()
```

The KDE plot highlights a clear relationship between game type and overall success, evident in the rightward shift and sharper peak of home wins. This difference underscores the impact of fan support and home-court familiarity in enhancing performance and ensuring greater consistency in game wins when playing at home.

Our next aim is to examine how the type of game (home vs. away) affects the relationship between wins and average attendance, so we separated the data into home and road games for analysis using a scatterplot. 

```python
sns.scatterplot(data = clean_dataset, x = clean_dataset['home_wins'], y = clean_dataset['home_avg_attendance'], label = 'Home')
sns.scatterplot(data = clean_dataset, x = clean_dataset['road_wins'], y = clean_dataset['road_avg_attendance'], label = 'Road')
plt.title('Wins vs Attendance from 2001-2023 by Type of Game')
plt.xlabel('Overall Wins')
plt.ylabel('Overall Average Attendance')
```

By examining the relationship between the dependent variables, types of attendance, and the independent variables, types of wins/elo, we aimed to determine the statistical significance of these relationships (using an alpha value of 0.05).

## Linear Relationship between Home Wins and Average Home Attendance

We can first take a look at the relationship between our independent variable of home wins having an effect on our dependent variable of average home attendance numbers using OLS regression.

```python
import patsy
import statsmodels.api as sm
df = clean_dataset[['home_wins','home_avg_attendance']]
df['home_avg_attendance'] = df['home_avg_attendance'] / 1000
df.columns = ['home_wins','home_avg_attendance']
outcome_1, pred_1 = patsy.dmatrices('home_avg_attendance~home_wins', df)
mod_1 = sm.OLS(outcome_1, pred_1)
res_1 = mod_1.fit()
print(res_1.summary())
```

To ensure that our analysis could meaningfully compare variables with drastically different scales, we scaled the attendance numbers down by a factor of 1,000. As a result, all reported outcomes for attendance are interpreted per 1,000 attendees. This normalization was necessary to maintain consistency in our regression model and to ensure the interpretability of our results.

The R-squared statistic from the OLS results show that about **5.9% of the variation in average home attendance is explained by the number of wins**, so the low number suggests that home wins is not a good predictor of average home attendance. Other factors not included in the model may play a more significant role in influencing attendance numbers.

Despite the low R-squared value, the **p-value** for the relationship between home wins and average home attendance is **less than 0.05**, indicating that this relationship is **statistically significant**.

## Linear Relationship between Road Wins and Average Road Attendance

Now we can look at the relationship between our independent variable of road wins and its effect on our dependent variable of average road attendance numbers using OLS regression.

```python
df = clean_dataset[['road_wins','road_avg_attendance']]
df['road_avg_attendance'] = df['road_avg_attendance'] / 1000
df.columns = ['road_wins','road_avg_attendance']
outcome_2, pred_2 = patsy.dmatrices('road_avg_attendance~road_wins', df)
mod_2 = sm.OLS(outcome_2, pred_2)
res_2 = mod_2.fit()
print(res_2.summary())
```

Similar to the results we got when we used home wins to predict home average attendance numbers, we got a very low R-squared value, 0.082, meaning that only **8.2% of the variation of road attendance numbers is affected by road wins**. However, we did receive a **p-value of less than 0.05**, meaning the relationship is still **statistically significant**.

## Linear Relationship between Overall Wins and Overall Attendance

Since we analyzed the relationships between our confounding variables of home/road wins and attendance numbers, we can move on to the overall wins and its effect as a predictor of overall average attendance numbers.

```python
df = clean_dataset[['overall_wins','overall_avg_attendance']]
df['overall_avg_attendance'] = df['overall_avg_attendance'] / 1000
df.columns = ['overall_wins','overall_avg_attendance']
outcome_2, pred_2 = patsy.dmatrices('overall_avg_attendance~overall_wins', df)
mod_2 = sm.OLS(outcome_2, pred_2)
res_2 = mod_2.fit()
print(res_2.summary())
```

We performed OLS regression on overall wins and its effect on overall average attendance, but we found that while their relationship is statistically significant due to their less than 0.05 p-value, **overall wins is not a strong indicator of the average attendance counts** due to their relatively low R-squared values, which turned out to be 0.097.

## Linear Relationship between ELO and Overall Attendance

Then we moved on from using wins as the performance metric to using average ELO of the teams to predict the overall attendance numbers.

```python
df = clean_dataset[['average_elo','overall_avg_attendance']]
df['overall_avg_attendance'] = df['overall_avg_attendance'] / 1000
df.columns = ['average_elo','overall_avg_attendance']
outcome_2, pred_2 = patsy.dmatrices('overall_avg_attendance~average_elo', df)
mod_2 = sm.OLS(outcome_2, pred_2)
res_2 = mod_2.fit()
print(res_2.summary())
```

Between the independent variable of average elo and dependent variable of overall average attendance, the increase in R-squared value of 0.267 indicates that about **26.7% of the variation of overall average attendance can be explained by ELO rating**, yet this statistic is still relatively low. But again the p-value is less than 0.05, showing statistically significance between the two variables.

## Linear Relationship between ELO + Overall Wins and Overall Attendance

Finally, we decided to use both ELO and overall wins as the predictors of overall attendance numbers to see if these factors together can explain the variation in attendance more effectively rather than using each predictor individually. We aim to determine their joint impact on attendance and assess the significance of each factor in driving fan engagement. 

```python
df = clean_dataset[['average_elo','overall_avg_attendance', 'overall_wins']]
df['overall_avg_attendance'] = df['overall_avg_attendance'] / 1000
df.columns = ['average_elo','overall_avg_attendance', 'overall_wins']
outcome_2, pred_2 = patsy.dmatrices('overall_avg_attendance~average_elo + overall_wins', df)
mod_2 = sm.OLS(outcome_2, pred_2)
res_2 = mod_2.fit()
print(res_2.summary())
```

The results showed that the **model explained 26.8% of the variation in overall average attendance**, suggesting that while these variables account for some of the variation in attendance, many other factors are still influencing attendance numbers. This was our highest R-squared value we got throughout our OLS regression analysis. 

We also got that our **average elo was statistically significant**, with a p-value of less than 0.05, but the **overall wins were not statistically significant** as their p-value of 0.469 was greater than 0.05.

## **Privacy**
We will only use publicly available aggregate data without personally identifiable information, such as ESPN's NBA Attendance Report, which do not contain any personally identifiable information. By doing so, we ensure that privacy risks are mitigated. However, it's important to note that ESPN's attendance data is intended for individual use, and scraping or republishing may violate its terms. ## **Bias**
Our analysis is subject to several potential biases inherent in the dataset and its collection:
- **Quality of Attendance**: Attendance figures do not differentiate between home and away fans. This may lead to inaccurate conclusions about fan loyalty to home teams.
- **Differences in Arena Capacity**: Different stadiums have varying capacities, which could skew comparisons of raw attendance figures.
- **Economic Disparities**: Cities hosting NBA teams differ in their economic conditions, which may influence attendance through variations of income and ticket prices.
- **Major Events**: Shared events, like the COVID-19 pandemic, and local events, like the opening of a new stadium, may significantly impact attendance, and could introduce confounding variables.

## **Fairness**
Our dataset and analysis could unintentionally favor certain teams or cities due to these biases:
- **Economic Inequities**: Teams in wealthier cities may naturally have higher attendance, independent of performance.
- **External Events**: Certain teams might experience attendance shifts due to unique events, like the opening of Chase Center for the Golden State Warriors in 2019-2020.
  
To address these biases, we plan to implement several strategies:
1. **Benchmarking Attendance**: To better compare attendance rates across the different teams, a team’s attendance could be measured by taking the proportion of attendees and maximum capacity for a team’s arena.
2. **Segmenting Data**: With major events like the COVID pandemic and the NBA lockout (2011-2012 season in which there was a shortage of games), certain seasons could be parsed out if broader factors limiting attendance are at play. We could possibly also look deeper into individual teams that experienced a local major event (such as the 2019-2020 Golden State Warriors, who played out of Chase Center for the first time) to determine if the event played a role in attendance rather than performance alone.
3. **Implementation of Control Variables**: By looking at economic indicators (unemployment rate, city population, cost of attendance, etc.), we could control for some external factors at play.

## **Transparency**
We will ensure transparency by clearly documenting our data sources, methods, and ethical considerations. Limitations and potential biases will be explicitly communicated, including their implications for our conclusions.

## **Mitigating Harm**
To mitigate potential harm:
- We will exclude outlier seasons or events when external factors heavily influence attendance.
- By segmenting data and benchmarking attendance, we aim to avoid misleading comparisons and ensure that our analysis reflects the true relationship between team performance and attendance.

In conclusion, while our project acknowledges these challenges, we have developed thoughtful strategies to address and mitigate them. This approach will enable us to deliver a robust and equitable analysis of NBA attendance trends.
